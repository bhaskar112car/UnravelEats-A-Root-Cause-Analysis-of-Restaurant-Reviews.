```python
# language detection and translation function
def lang_trans(data: 'data_column'):
    # detect the language of the data
    lang = detect(data)
    # if the language is English, return the data as it is
    if lang == 'en':
        return data
    # otherwise, use GoogleTranslator to translate the data to English
    else:
        # create a GoogleTranslator object with auto source and English target
        google_translator = GoogleTranslator(source='auto', target='en')
        # translate the data and return it
        translated_data = google_translator.translate(data)
        return translated_data
```

```python
def remove_lines(data):
    clean_text = data.replace("\n",' ').replace('\t',' ')
    return clean_text

# contraction mapping
def expand_text(data):
    expanded_doc=contractions.fix(data)
    return expanded_doc

# handled accented charactores
def accented_char(data):
    fixed_text=unidecode(data)
    return fixed_text
```

```python
# import necessary library
from nltk.corpus import stopwords

# creating stopword list
stopword_list = stopwords.words('english')
stopword_list.remove('not')
stopword_list.remove('no')
stopword_list.remove('nor')
stopword_list
```

```python
# clean data
def clean_data(data):
    tokens=word_tokenize(data)
    normalization=[word.lower() for word in tokens]
    remove_punc = [word for word in normalization if word not in punctuation]
    words_without_stop = [word for word in remove_punc if word not in stopword_list ]
    clean_text = [word for word in words_without_stop if len(word)>2]
    return clean_text
# lemmatization
def lemmatization(data):
    lemmatizer = WordNetLemmatizer()
    final_text = []
    for word in data :
        lemmatized_word = lemmatizer.lemmatize(word)
        final_text.append(lemmatized_word)
        return final_text
    
```

```python
from textblob import TextBlob

def find_negative_reviews(review):
    text = TextBlob(review)
    polarity = text.sentiment.polarity
    if polarity < 0:
        return 0
    else:
        return 1
```

```python
revdf['Lable']=revdf['Review'].apply(find_negative_reviews)
revdf
```

```python
clean_text_train=x_train.apply(remove_lines)
clean_text_test=x_test.apply(remove_lines)

clean_text_train=clean_text_train.apply(expand_text)
clean_text_test=clean_text_test.apply(expand_text)

clean_text_train=clean_text_train.apply(accented_char)
clean_text_test=clean_text_test.apply(accented_char)

clean_text_train=clean_text_train.apply(clean_data)
clean_text_test=clean_text_test.apply(clean_data)

clean_text_train=clean_text_train.apply(lemmatization)
clean_text_test=clean_text_test.apply(lemmatization)
```

# count vectorizer

```python
count= CountVectorizer(max_df=0.98,max_features=1000)
# max_df : 0.95 : 1 will not create features for word came more than 95% of 100 documents
# min_df : 15 : 100(100 doc df) we will not create featurs for words which less thAn 15
# max_features : no of featurs to create in countvectorizer
# stopwords : 'english' optional is not done in pre-processing
# lowercase : optional if forgot in pre-proc
```

```python
count_train_val=count.fit_transform(clean_text_train)
count_test_val=count.transform(clean_text_test)
```
